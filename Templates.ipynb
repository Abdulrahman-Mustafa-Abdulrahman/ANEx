We'll first create a DataFrame of only the demographic properties of our sample. This will be used for some exploration of said properties:

- Code: Patient codes consisting of three parts:

1. Decade of Age: First digit of Age (Next cell).

undefined. Education: Education Code (Cell After Next). 

undefined. Index: Three-digit numerical index.

Education	Education Code (Cell After Next). 

- Education: Code for education as below:

1. No Education: 0.

undefined. Primary School: 1.

undefined. Preparatory School: 2.

undefined. Secondary School: 3.

undefined. Diploma: 4.

undefined. Institution: 5.

undefined. University: 6.

- GA: Gestational age of current pregnancy in months.

- Primigravida: Whether it's the patient's first pregnancy or not:

1. Primigravida: 1.

undefined. Not primigravida: 0.

- L: Number of living children at the time of conducting the questionnaire.

- A: Number of previous aborions at the time of conducting the questionnaire.

- P: Number of pre-term deliveries at the time of conducting the questionnaire.

- F: Number of full-term deliveries at the time of conducting the questionnaire.

- VD: Number of previous vaginal deliveries at the time of conducting the questionnaire.

- CS: Number of previous cesarean sections the time of conducting the questionnaire.

- AAP: Previous medical advice against practice.

- History of Practice: Previous history of exercise before pregnancy.

- Age: Age in years.

```python
fig = px.pie(values=df.Primigravida.value_counts(),
             names=['Non-Primigravida','Primigravida'],
             height=1000, 
             width=1000,
             color_discrete_sequence=['#D9D9D9','#843C0C'])

fig.update_traces(#title=dict(text='Distribution of Sth Against Sth in our dataset',
                  #           font=dict(size=28),
                  #           position='top center'),
                  textposition='outside',
    
                  textfont=dict(size=50,
                                color='black'),
    
                  showlegend=False)

fig.update_layout(font=dict(family='Arial'),
                  plot_bgcolor='rgba(0,0,0,0)',
                  paper_bgcolor='rgba(0,0,0,0)')

fig.show()
```
